```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prop = pd.read_csv("properatti.csv", sep = ",")
prop.shape
```

```python
prop.rename(columns={'Unnamed: 0':'ID'}, inplace=True)
```

```python
prop.columns
```

```python
prop.drop(prop.columns[[1,4,5,7,8,13,14,17,18,22,25]], axis=1, inplace=True)
```

```python
prop['Superficie_No_Cubierta'] = prop['surface_total_in_m2'] - prop['surface_covered_in_m2']
```

```python
prop['Precio_M2_Superficie_Total'] = prop['price'] / prop['surface_total_in_m2']
```

```python
prop['Precio_M2_Superficie_Cubierta'] = prop['Precio_M2_Superficie_Total'] * prop['surface_covered_in_m2']
```

```python
prop['Precio_M2_Superficie_No_Cubierta'] = prop['Precio_M2_Superficie_Total'] * prop['Superficie_No_Cubierta']
```
